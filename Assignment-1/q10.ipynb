```python
#QUESTION 10
import matplotlib.pyplot as plt
male_age = [53,51,71,31,33,39,52,27,54,30,64,26,21,54,52,20,59,32]
female_age = [53,65,68,21,75,46,24,63,61,24,49,41,39,40,25,54,42,32,48,23,23]
# Plot histogram for male age
plt.hist(male_age, bins=10, color='blue', alpha=0.5)
plt.title('Histogram of Male Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot histogram for female age
plt.hist(female_age, bins=10, color='red', alpha=0.5)
plt.title('Histogram of Female Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
