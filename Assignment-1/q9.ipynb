```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10,100)
d1 = x*np.sin(x)
d2 = x**2*np.sin(x)
d3 = x**3*np.sin(x)
d4 = x**4*np.sin(x)

print(d1)
```

```python
plt.plot(x,d1, label='xsin(x)')
plt.plot(x,d2, label='x^2sin(x)')
plt.plot(x,d3, label='x^3sin(x)')
plt.plot(x,d4, label='x^4sin(x)')

plt.title('sine functions plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show();
```
