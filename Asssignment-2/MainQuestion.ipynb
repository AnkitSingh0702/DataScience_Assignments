1. Write a menu-driven program to perform Addition, Subtraction, Scalar Multiplication, Dot Product
and Length of vectors.

```python
#Answer-1
import math

def add_vectors(vec1, vec2):
    """Adds two vectors."""
    return [a + b for a, b in zip(vec1, vec2)]

def subtract_vectors(vec1, vec2):
    """Subtracts two vectors."""
    return [a - b for a, b in zip(vec1, vec2)]

def multiply_vector_by_scalar(vec, scalar):
    """Multiplies a vector by a scalar."""
    return [a * scalar for a in vec]

def dot_product(vec1, vec2):
    """Calculates the dot product of two vectors."""
    return sum(a * b for a, b in zip(vec1, vec2))

def vector_length(vec):
    """Calculates the length of a vector."""
    return math.sqrt(sum(a**2 for a in vec))

def get_vector():
    """Prompts the user to enter a vector."""
    vec_str = input("Enter a vector in the format 'x,y,z': ")
    vec = [float(a.strip()) for a in vec_str.split(",")]
    return vec

def print_vector(vec):
    """Prints a vector."""
    print("[" + ", ".join(str(a) for a in vec) + "]")

def main():
    while True:
        print("Menu:")
        print("1. Add vectors")
        print("2. Subtract vectors")
        print("3. Multiply vector by scalar")
        print("4. Calculate dot product")
        print("5. Calculate vector length")
        print("6. Quit")
        choice = input("Enter your choice (1-6): ")

        if choice == "1":
            vec1 = get_vector()
            vec2 = get_vector()
            result = add_vectors(vec1, vec2)
            print("The sum of the vectors is:")
            print_vector(result)
        elif choice == "2":
            vec1 = get_vector()
            vec2 = get_vector()
            result = subtract_vectors(vec1, vec2)
            print("The difference of the vectors is:")
            print_vector(result)
        elif choice == "3":
            vec = get_vector()
            scalar = float(input("Enter a scalar: "))
            result = multiply_vector_by_scalar(vec, scalar)
            print("The product of the vector and scalar is:")
            print_vector(result)
        elif choice == "4":
            vec1 = get_vector()
            vec2 = get_vector()
            result = dot_product(vec1, vec2)
            print("The dot product of the vectors is:", result)
        elif choice == "5":
            vec = get_vector()
            result = vector_length(vec)
            print("The length of the vector is:", result)
        elif choice == "6":
            break
        else:
            print("Invalid choice. Please enter a number from 1 to 6.")

if __name__ == "__main__":
    main()
```

2. Write a program that takes the order of the matrix and creates a matrix in the following manner:The (ij)th entry of the matrix should be the sum of i and j. Eg: The 0th row and 0th column should have the value (0+0) i.e. 0 and the 0 th row and first column should have value (0+1) i.e. 1 and so on.


```python
#Answer-2
def create_matrix(n):
    matrix = [[i + j for j in range(n)] for i in range(n)]
    return matrix

def print_matrix(matrix):
    for row in matrix:
        print(" ".join(str(x) for x in row))

def main():
    n = int(input("Enter the order of the matrix: "))
    matrix = create_matrix(n)
    print("The matrix is:")
    print_matrix(matrix)

if __name__ == "__main__":
    main()

```

3. Write two functions that extract the rows and columns of a matrix A.

```python
def get_rows(matrix):
    rows = []
    for row in matrix:
        rows.append(row)
    return rows

def get_columns(matrix):
    columns = []
    num_rows = len(matrix)
    num_columns = len(matrix[0]) if num_rows > 0 else 0

    for j in range(num_columns):
        column = []
        for i in range(num_rows):
            column.append(matrix[i][j])
        columns.append(column)
    return columns
```

4. Write a function to compute the component-wise mean of a list of vectors. Assert the condition that
the vectors must be of same length.

```python
def mean(vectors):
    assert all(len(v) == len(vectors[0]) for v in vectors)
    num = len(vectors)
    length_vector = len(vectors[0])
    vector_mean = [0] * length_vector
    
    for i in range(vector_length):
        totalsum = sum(v[i] for v in vectors)
        vector_mean[i] = totalsum / num
    
    return vector_mean
```

6. Write a program to find median of a given list of integers. Combine both odd and even number of
terms

```python
def median(nums):
    sort = sorted(nums)
    length = len(sort)

    if length % 2 == 0:
        right = length // 2
        left = right - 1
        median = (sort[left] + sort[right]) / 2
    else:
        mid = length // 2
        median = sort[mid]
    return median

numbers = [9,5,7,2,1,8,6,4,3]
print(median(numbers))
```

8. Plot the Normal PDFs using various value of µ and σ

```python
import math
import matplotlib.pyplot as plt

def normal_pdf(x, mu=0, sigma=1):
    sqrt_two_pi = math.sqrt(2 * math.pi)
    return (math.exp(-(x - mu) ** 2 / (2 * sigma ** 2)) / (sqrt_two_pi * sigma))

xs = [x for x in range(-10, 10)]
plt.plot(xs, [normal_pdf(x, 0, 1) for x in xs], '-', label='mu=0, sigma=1')
plt.plot(xs, [normal_pdf(x, 0, 2) for x in xs], '.', label='mu=0, sigma=2')
plt.plot(xs, [normal_pdf(x, 0, 0.5) for x in xs], '-.', label='mu=0, sigma=0.5')
plt.plot(xs, [normal_pdf(x, -1, 1) for x in xs], ':', label='mu=-1, sigma=1')
plt.legend()
plt.show()
```

9. Do the same as above question for Normal CDFs using the same values of µ and σ.

```python
import math
import matplotlib.pyplot as plt
def normal_cdf(x, mu=0, sigma=1):
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2
xs = [x for x in range(-10, 10)]
plt.plot(xs, [normal_cdf(x, 0, 1) for x in xs], '-', label='mu=0, sigma=1')
plt.plot(xs, [normal_cdf(x, 0, 2) for x in xs], '.', label='mu=0, sigma=2')
plt.plot(xs, [normal_cdf(x, 0, 0.5) for x in xs], '-.', label='mu=0, sigma=0.5')
plt.plot(xs, [normal_cdf(x, -1, 1) for x in xs], ':', label='mu=-1, sigma=1')
plt.legend()
plt.show()
```

12. Using the Binomial(n, p) distribution plot a histogram to show the actual binomial samples. Plot both in the same graph.

```python
import random
from collections import Counter

def bernoulli_trial(p):
    return 1 if random.random() < p else 0

def binomial(n, p):
    return sum(bernoulli_trial(p) for _ in range(n))

def make_hist(p, n, num_points):
    data = [binomial(n, p) for _ in range(num_points)]
# use a bar chart to show the actual binomial samples
    histogram = Counter(data)
    plt.bar([x - 0.4 for x in histogram.keys()],
    [v / num_points for v in histogram.values()],
    0.8,
    color='0.75')
    mu = p * n
    sigma = math.sqrt(n * p * (1 - p))
# use a line chart to show the normal approximation
    xs = range(min(data), max(data) + 1)
    ys = [normal_cdf(i + 0.5, mu, sigma) - normal_cdf(i - 0.5, mu, sigma)
    for i in xs]
    plt.plot(xs,ys)
    plt.title("Binomial Distribution vs. Normal Approximation")
    plt.show()
make_hist(0.75, 100, 100)
```
