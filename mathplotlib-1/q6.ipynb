```python
import matplotlib.pyplot as plt

ax = plt.subplot()
fruits = ["Apple" , "Orange" ,  "Cheery ","Bluebery"]
counts = [40,100,30,55]
bar_labels = ['red','blue', 'red', 'orange']
bar_color= ['tab:red','tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits,counts,label=bar_labels,color = bar_color)
ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind color')
ax.legend(title = 'Fruit color')
```
