### Q1: Approximating derivative

```python
import math
import matplotlib.pyplot as plt


def f1(x):
  return x**2
def difference_quotient(f, x, h=0.001):

    return (f(x + h) - f(x)) / h
difference_quotient(f1,2,0.001)
x=[i/10 for i in range(-50,50)]

y=[difference_quotient(f1,i,0.00001) for i in x]
y1=[2*i for i in x]
print(y)
print(y1)

plt.plot(x,y,'rx',label='approximation')
plt.plot(x,y1,'b+',label='original')
plt.legend()
plt.show()
```

### Approximating the partial derivative

```python
def f1(l):
  return (l[0]**2)*l[1]
 
def partial_difference_quotient(f, v, i, h):
    
    w = [v_j + (h if j == i else 0)
         for j, v_j in enumerate(v)]
    return (f(w) - f(v)) / h
def estimate_gradient(f, v, h=0.0001):
    return [partial_difference_quotient(f, v, i, h) 
            for i in range(len(v))]
estimate_gradient(f1, [2,3])
```

### Gradient descent

```python
#minimise x square using gradient descent concept

def derivative_of_x_square(x):
  return 2*x
target_x=10
learning_rate=0.1
iterations=5
for i in range(iterations):
  target_x=target_x-(derivative_of_x_square(target_x)*learning_rate)
print(target_x)


### Minimise function of two variable using gradient descent

def derivative_of_x_square_plus_y_square(x,y):
  return [2*x,2*y]
target_x,target_y=10,10
learning_rate=0.1
iterations=5
for i in range(iterations):
  target_x=target_x-(2*target_x*learning_rate)
  target_y=target_y-(2*target_y*learning_rate)
  print(i,target_x,target_y)


# Minimise Sum of squares

def step(v, gradient, LearningRate):
    return [v_i + LearningRate * gradient_i
            for v_i, gradient_i in zip(v, gradient)]
def sum_of_squares_gradient(v):
    return [2 * v_i for v_i in v]
v=[1,1,1]
for i in range(10):
  gradient=sum_of_squares_gradient(v)
  v=step(v,gradient,-0.01)
  print(i,v)
```

### Linear regression


```python
import matplotlib.pyplot as plt

# Generate some sample data
x = [1,2]
y = [10,7]

# Define the slope and intercept of the line
m = 0
b = 0

# Set the learning rate and number of iterations
alpha = 0.01
iterations = 1000

# Define the gradient descent algorithm
n = float(len(x))
for i in range(iterations):
    y_pred = [(m*x[i]+b) for i in range(len(x))]
    error = [y_pred[i] - y[i] for i in range(len(y))]
    m = m - alpha * (1/n) * sum([error[i]*x[i] for i in range(len(x))])
    b = b - alpha * (1/n) * sum(error)

# Print the final slope and intercept
print("Slope: ", m)
print("Intercept: ", b)

# Plot the data and regression line
plt.scatter(x, y)
plt.plot(x, [(m*x[i]+b) for i in range(len(x))])
plt.show()
```
