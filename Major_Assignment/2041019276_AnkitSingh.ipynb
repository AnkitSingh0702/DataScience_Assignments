### 1: Write a python program to group the user ids corresponding to the following age group.
#### • less or equal to 20.
#### • between 20 and 40.
#### • more than 40.

```python
less_or_equal_20 = []
between_20_and_40 = []
more_than_40 = []

for d in data:
    if d["Age"] <= 20:
        less_or_equal_20.append(d["user_id"])
    elif d["Age"] > 20 and d["Age"] <= 40:
        between_20_and_40.append(d["user_id"])
    else:
        more_than_40.append(d["user_id"])

# Print the results
print("User IDs less than or equal to 20:", less_or_equal_20)
print("User IDs between 20 and 40:", between_20_and_40)
print("User IDs more than 40:", more_than_40)
```

### 3. Write a python program to create a list of 20 vectors by taking user’s height, weight and age. Find the sum of all these 20 vectors. After that, find the average height, average weight and average age. Hint: [150,60,23] will be one vector.

```python
vectors =[]
for user in data:
    temp=[user["Height"],user["Weight"],user["Age"]]
    vectors.append(temp);

height_sum=0
weight_sum=0
age_sum=0

for e in vectors:
    height_sum+=e[0]
    weight_sum+=e[1]
    age_sum+=e[2]

print("Average Height --> ",height_sum/len(vectors))
print("Average Weight --> ",weight_sum/len(vectors))
print("Average Age --> ",age_sum/len(vectors))
```

### 4. Write a python program to find the mean, median and mode of weights of users.

```python
#mean
def mean(d):
    return sum(d)/len(d)

#median
def median(d):
    data_sorted = sorted(d)
    n = len(data_sorted)
    mid = n//2

    if n%2 == 0:
        return (data_sorted[mid - 1] + data_sorted[mid]) / 2
    else:
        return data_sorted[mid]

#mode   
def mode(d):
    frq = {}
    for i in d:
        if i not in frq:
            frq[i] = 1
        else:
            frq[i] += 1
    for k in frq.keys():
        if frq[k] == max(frq.values()):
            return k

print("Mean weight-->",mean(weight))
print("Median weight--> ",median(weight))
print("Mode weight--> ",mode(weight))
```

### 5. Write a python program to find the variance of heights of users.

```python
heights = [d["Height"] for d in data]
mean_height = sum(heights) / len(heights)

# Calculate the variance of the heights
variance = sum((h - mean_height) ** 2 for h in heights) / len(heights)

print("Variance of heights:", variance)
```

### 6. Write a python program to find covariance between weights and heights of users.

```python
n = len(data)

def covariance(h,w):
    h_mean = mean(h)
    w_mean = mean(w)
    return sum([(h[i]-h_mean)*(w[i]-w_mean) for i in range(n)])/n

print("Covariance between heights and weights -->", covariance(heights,weight))
```
